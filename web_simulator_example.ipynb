```python
from web_simulator import web_utils
```

```python
N = 1000
func = lambda x, y: (x ** 2 + y ** 2) / 2
threshold_min = 0.001
threshold_max = 2.0

web = web_utils.sample_web(func, N=N, threshold_min=threshold_min, threshold_max=threshold_max)
print(f"Number of edges: {web.num_edges}")

df = web.to_df()
print(df.iloc[:10])
```

```python
# plot 
web.plot()
```

```python
# Save as csv files
path = "./synthetic_web"
web.save(path)
```
